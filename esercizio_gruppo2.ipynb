### **Calcolo della pressione media in output $\langle P \rangle$**


### **Confronto tra l'equazione di stato dei gas ideali e quella di van der Waals**

$P_{ideale} = \frac{nRT}{V}$



$P_{vdW} = \frac{nRT - \frac{an^2}{V} + \frac{abn^3}{V^2}}{V-nb}$



```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Dati della simulazione in input (variabili termodinamiche)
N = 1000  # numero di particelle
T = 523   # temperatura
R = 8.31  # costante dei gas
N_A = 6.022*10**23 # Numero di Avogadro
n = N/N_A # Calcolo numero di moli

# Dati della simulazione in output
p_simulazione = np.array([6.42,4.98,3.94,3.17,2.16])
# output delle simulazioni a diverse densit√† (restringendo il volume della
# scatola mantenendo cost il numero di particelle) (in atm)
p_simulazione = p_simulazione*10**5  # in Pa
L_simulazione = np.array([220,240,260,280,319])*10**(-10) # in m
V_simulazione = L_simulazione**3 #definisco i volumi come i cubi dei lati della scatola

# Vettori per plot
L = np.linspace(200*10**(-10),330*10**(-10),10000) # in m
V = L**3 # in m^3

# Parametri dell'equazione di van der Waals per la CO2
a = 1.824 #(J*m^3)/(mol^2)
b = 0.0001154 #m^3/mol

# Equazioni di stato ideale e van der waals esplicitate nella pressione
p_vdW = (n*R*T - (a*n**2)/V + (a*b*n**3)/V**2)/(V-n*b)
p_ideale = (n*R*T)/V

# Isoterma in scala lineare
plt.figure(1)
plt.plot(L_simulazione,p_simulazione, 'o')  # i pallini sono i dati delle simulazioni
plt.plot(L,p_vdW, 'black')  # la linea nera √® la previsione per van der waals
plt.plot(L,p_ideale, 'orange')  # la linea arancione √® la previsione per l'eq. dei gas ideali
plt.xlim(200*10**(-10),330*10**(-10))
plt.ylim(0,0.1*10**(7))
plt.xlabel('L [m]')
plt.ylabel('P [Pa]')
plt.title('P vs L - scala lineare')

# Isoterma in scala logaritmica
plt.figure(2)
plt.loglog(L_simulazione,p_simulazione, 'o')
plt.loglog(L,p_vdW, 'black')
plt.loglog(L,p_ideale, 'orange')
plt.xlim(200*10**(-10),330*10**(-10))
plt.ylim(10**5,0.1*10**(7))
plt.xlabel('L [m]')
plt.ylabel('P [Pa]')
plt.title('P vs L - scala logaritmica')

# Scostamento percentuale dall'idealit√†
plt.figure(3)
plt.plot(L,abs((p_ideale-p_vdW)/(p_vdW))*100)
plt.xlim(200*10**(-10),330*10**(-10))
plt.ylim(0,10)
plt.xlabel('L [m]')
plt.ylabel('Scostamento %')
plt.title('Scostamento pressione VdW e ideale')
```

### **Dinamica nella fase di produzione**
$ D = \lim_{t->+\infty} \frac{1}{6t}MSD$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Caricamento dei file
MSD_file = np.genfromtxt('msd_319.dat', delimiter=' ')
time = MSD_file[:, 0]
msd = MSD_file[:, 1]

# Plot MSD contro il tempo in scala lineare
plt.figure(1)
plt.xlabel("t [fs]")
plt.ylabel("MSD [Angstrom^2]")
plt.title("MSD vs Tempo (Scala lineare)")
plt.plot(time,msd)

# Plot MSD contro il tempo in scala logaritmica
plt.figure(2) 
plt.loglog(time,msd, '-')
plt.xlabel("t [fs]")
plt.ylabel("MSD [Angstrom^2]")
plt.title("MSD vs Tempo (Scala logaritmica)")

# Definizione della funzione che useremo per fittare i dati
def linea_retta(t,D):
  return 6*D*t

# Fit dei dati
D, err = curve_fit(linea_retta,time[10001:],msd[10001:])

# Plot MSD contro il tempo in scala logaritmica con fit
plt.figure(3)
plt.loglog(time,msd)
plt.plot(time,linea_retta(time,D), '--')
plt.xlabel("t [fs]")
plt.ylabel("MSD [Angstrom^2]")
plt.title("MSD vs Tempo (Scala logaritmica)")

print(D)
```
